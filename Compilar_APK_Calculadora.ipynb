## ✅ VERIFICACIÓN: Archivos Necesarios

**Ejecuta esta celda para verificar que todo está listo antes de compilar**

````python
import os

print("🔍 Verificando archivos necesarios...\n")

# Lista de archivos requeridos
archivos_requeridos = {
    'calculadora_android.py': 'Código principal de la app',
    'buildozer.spec': 'Configuración de compilación'
}

errores = []
advertencias = []

# Verificar cada archivo
for archivo, descripcion in archivos_requeridos.items():
    if os.path.exists(archivo):
        size = os.path.getsize(archivo)
        print(f"✅ {archivo} - {descripcion}")
        print(f"   Tamaño: {size:,} bytes")
        
        # Verificar que no esté vacío
        if size == 0:
            errores.append(f"❌ {archivo} está vacío")
    else:
        errores.append(f"❌ Falta: {archivo} - {descripcion}")

# Verificar archivo opcional
if os.path.exists('icono.ico'):
    print(f"✅ icono.ico - Ícono de la app (opcional)")
else:
    advertencias.append("⚠️ No se encontró icono.ico (se usará ícono por defecto)")

# Verificar permisos
print("\n📋 Permisos de archivos:")
!ls -lh calculadora_android.py buildozer.spec 2>/dev/null

# Verificar contenido de buildozer.spec
print("\n📄 Primeras líneas de buildozer.spec:")
!head -5 buildozer.spec

print("\n" + "="*50)
if errores:
    print("❌ ERRORES ENCONTRADOS:")
    for error in errores:
        print(f"   {error}")
    print("\n⛔ NO PUEDES CONTINUAR. Sube los archivos faltantes.")
elif advertencias:
    print("⚠️ ADVERTENCIAS:")
    for adv in advertencias:
        print(f"   {adv}")
    print("\n✅ Puedes continuar, pero revisa las advertencias.")
else:
    print("✅ TODO LISTO PARA COMPILAR")
    print("👉 Continúa con el siguiente paso")
print("="*50)
````

## 🚀 PASO 4: Compilar el APK

**⚠️ IMPORTANTE:**
- **Primera vez:** Tardará 30-60 minutos
- **No cierres esta pestaña** mientras compila
- Verás muchas líneas de texto, es normal

**☕ Ve por un café, esto tardará un rato...**

````python
import sys
import subprocess

print("🚀 Iniciando compilación del APK...")
print("⏱️ Primera vez: 30-60 minutos")
print("⏱️ Siguientes veces: 5-10 minutos")
print("\n📥 Descargando Android SDK y NDK...")
print("\n" + "="*50)

# Ejecutar buildozer y capturar el resultado
result = subprocess.run(
    ['buildozer', 'android', 'debug'],
    capture_output=False,
    text=True
)

print("\n" + "="*50)

if result.returncode == 0:
    print("✅ ¡Compilación completada exitosamente!")
    print("\n🎉 APK generado en: bin/calculadoraprofesional-2.0-debug.apk")
else:
    print("❌ ERROR: La compilación falló")
    print(f"❌ Código de error: {result.returncode}")
    print("\n⚠️ EJECUTA LA SIGUIENTE CELDA para diagnosticar el problema")
````

## 🔍 DIAGNÓSTICO: Si hubo error en la compilación

**Ejecuta esta celda SOLO si el paso anterior falló**

Esto te mostrará el error real que ocurrió.

````python
print("🔍 Diagnosticando el error...\n")

# Ver si existe el directorio .buildozer
print("1️⃣ Verificando directorio .buildozer:")
!ls -la .buildozer 2>/dev/null || echo "❌ No existe .buildozer (buildozer no se ejecutó)"

# Ver logs si existen
print("\n2️⃣ Buscando logs de error:")
!find .buildozer -name "*.log" -type f 2>/dev/null | head -5

# Mostrar últimas líneas del log principal si existe
print("\n3️⃣ Últimas líneas del log principal:")
!tail -50 .buildozer/android/platform/build-*/build.log 2>/dev/null || echo "❌ No hay logs disponibles"

# Verificar buildozer.spec
print("\n4️⃣ Verificando buildozer.spec:")
!head -20 buildozer.spec

# Verificar permisos
print("\n5️⃣ Verificando permisos:")
!ls -lh calculadora_android.py buildozer.spec

print("\n" + "="*50)
print("💡 POSIBLES CAUSAS DEL ERROR:")
print("="*50)
print("1. ❌ Falta el archivo calculadora_android.py")
print("2. ❌ buildozer.spec mal configurado")
print("3. ❌ Dependencias no instaladas correctamente")
print("4. ❌ Problemas de conexión al descargar SDK/NDK")
print("5. ❌ Error en el código Python")
print("\n👉 Revisa los mensajes arriba para identificar el problema")
````
